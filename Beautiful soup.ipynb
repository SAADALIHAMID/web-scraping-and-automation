```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

# Step 1: Fetch the webpage
url = "https://ticker.finology.in/"
r = requests.get(url)
soup = bs(r.text, "lxml")

# Step 2: Locate the table
table = soup.find("table", class_="table table-sm table-hover screenertable")

# Step 3: Extract table headers
headers = table.find_all("th")
titles = [i.text.strip() for i in headers]  # Remove whitespace

# Create an empty DataFrame with the extracted headers
df = pd.DataFrame(columns=titles)

# Step 4: Extract table rows
rows = table.find_all("tr")

# Step 5: Iterate over rows to fill the DataFrame
for i in rows[1:]:  # Skip the header row
    data = i.find_all("td")
    row = [tr.text.strip() for tr in data]  # Remove whitespace
    l = len(df)
    df.loc[l] = row  # Append the row to the DataFrame

# Display the DataFrame
print(df)

# Step 6: Convert DataFrame to CSV and remove any backslashes or unwanted characters
df.to_csv("finology_data.csv", index=False)

# Optional: Display the download link for Jupyter Notebook
from IPython.display import FileLink
display(FileLink('finology_data.csv'))
```

```python
import requests
from bs4 import BeautifulSoup as bs
url = "https://www.tutorialsfreak.com/tutorials"
web = requests.get(url)
soup=bs(web.content,"html.parser")
# print(soup.prettify())
```

```python
lines=soup.find_all("p")
```

```python
lines
```

```python
for i in lines:
    print(i.text)
```

```python
par=soup.find_all("div",class_="fs-16 fw-400 lh-24 label-color-1 mb-3 card-text")
par
```

```python
for i in par:
    print(i.text)
```

```python
a=soup.find("p",class_="fs-16 fw-400 lh-24 label-color-1 mb-3 card-text")
for i in a:
    print(i.text)
```

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```

```python
img=soup.find_all("img")
img
```

```python
for i in img:
    print(i.get("src"))
```
